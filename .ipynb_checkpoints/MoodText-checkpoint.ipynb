```python
from nltk.text import Text
```

```python
x=Text(words)
```

```python
x.concordance("月", width=10, lines=25)
```

```python
x.common_contexts("月")
```

```python
x.similar("店")
```

```python
x.count("月")
```

```python
x.vocab()
```

```python
import nltk
import jieba

raw=words
text=nltk.text.Text(jieba.lcut(raw))
```

```python
print(text.concordance('一個'))
```

```python
print(text.common_contexts(['一起']))
```

```python
text.collocations()
```

```python
text.dispersion_plot(['一個','東西'])
```

```python
fdist = nltk.FreqDist(nltk.word_tokenize(raw))
```

```python
thu1 = thulac.thulac(seg_only=True)  #只进行分词，不进行词性标注
thu1.cut(words,text=True) #对input.txt文件内容进行分词，输出到output.txt
```

```python
thu1 = thulac.thulac(seg_only=True,filt=True)  #只进行分词，不进行词性标注
thu1.cut(words,text=True) #对input.txt文件内容进行分词，输出到output.txt
```
