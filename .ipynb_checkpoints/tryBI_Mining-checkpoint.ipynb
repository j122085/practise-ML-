```python
import pymssql
conn = pymssql.connect(server='datacenter.wowprime.com', user='python',
                       password='12321', database='WangGroup') 
cursor = conn.cursor()
cursor.execute("SELECT [Called],[CareerName],[StoreNo],[StoreName],[Corporation_ch],[CloseDate] \
    FROM [WangGroup].[dbo].[StoreDetail_View]\
    WHERE [CloseDate] is NULL\
    AND right(rtrim(StoreNo),2) <> '00'\
    AND LEFT(rtrim(Called),2) <> '大陸';")  
wowprimeDienData = cursor.fetchall()
len(wowprimeDienData[0])
```

```python
cols=["Called","CareerName","StoreNo","StoreName","Corporation_ch","CloseDate"]
```

```python
conn.close()
```

```python
wowprimeDienDatalist=[]
for dien in wowprimeDienData:
    diendict={}
    for c,d in zip(cols,dien):
        if type(d)==str:
            d=d.strip()
        elif type(d)!=int:
            d=str(d)
        diendict[c]=d
    wowprimeDienDatalist.append(diendict)
```

```python
vlookName={dien['StoreNo']:dien['Called'] for dien in wowprimeDienDatalist}
vlookOtherName={dien['StoreNo']:dien['Corporation_ch'] for dien in wowprimeDienDatalist}
```
